## Exercise 1 – How long are movie titles?

In this exercise, you will apply descriptive statistics techniques to a real dataset: a database of 1000 popular movies from the IMDb website.

From the title of each movie, you will:
- Calculate the title length (in characters).
- Obtain key statistical metrics: mean, median, mode, range, variance, standard deviation, skewness, and kurtosis.
- Identify the longest and shortest titles.
- Visualize the distribution of these lengths with a histogram and a density curve.


```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, mode

# Load IMDb dataset
url = "../assets/imdb_1000.csv"
df = pd.read_csv(url)
df
```

```python
# Create a new column called 'title_length' with the length (number of characters) of each title

df['title_length'] = df['title'].str.len()

#df.head()

# More difficult version

# create the new column with a default value of 0
#df['title_length_1'] = 0
#df.head()
# check to make sure we only have a single row per title
#df['title'].value_counts

#for i in range(len(df)):
    #update the title_length for the row where title = title
    #title = df['title'][i]
    #title_length = len(title)
    #df['title_length_1'][i] = title_length


#df.head()
```

```python
# Calculate mean, median, and mode of the 'title_length' column
# mean
mean_of_title_length = df['title_length'].mean()
median_of_title_length = df['title_length'].median()
mode_of_title_length = df['title_length'].mode()[0]
```

```python
# Calculate the range, variance, and standard deviation of 'title_length'
# range is the difference between the maximum and minimum values
range_of_title_length = df['title_length'].max() - df['title_length'].min()
#range_of_title_length

# variance (less desireable) (squared standard deviation) is how far on average the values are from the mean.
variance_of_title_length = df['title_length'].var()
#print(variance_of_title_length)

# standard deviation (more desireable) (square root of variance) is also how far on average the values are from the mean.
std_of_title_length = df['title_length'].std()
#std_of_title_length 

```

```python
# Calculate skewness and kurtosis of the 'title_length' column
skewness_title_length = df['title_length'].skew()
print(skewness_title_length)
print('A positive skew tells us the data is right skewed, which we can verify with a histogram.')
plt.hist(df['title_length'], bins=30)
plt.show()
```

```python
# Print the shortest and longest title based on their length
print(df['title'][df['title_length'].idxmin()])
print(df['title'][df['title_length'].idxmax()])
```

```python
#plot your results

plt.hist(df['title_length'], bins=20, edgecolor='black')
plt.axvline(mean_of_title_length, color='red', label=f"Mean: {mean_of_title_length:.2f}")
plt.axvline(median_of_title_length, color='yellow', label=f"Median: {median_of_title_length:.2f}")
plt.axvline(mode_of_title_length, color='purple', label=f"Mode: {mode_of_title_length:.2f}")
plt.title('Distribution of Title Length with Central Tendencies')
plt.xlabel('Length of Title (Number of Characters)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

## Exercise 2 – Manual Calculation of Standard Deviation

In this exercise, you will manually calculate the standard deviation by selecting five actual titles from the movie dataset you loaded earlier.


```python
# Create a list with 5 actual values from df['title_length'], for example: [10, 13, 14, 18, 22]
import random as random

rand_title_lengths = random.sample(list(df['title_length']), 5)

print(rand_title_lengths)
```

```python
rand_title_lengths = pd.Series(rand_title_lengths)
```

```python
# Calculate the mean of the 5 values
mu = rand_title_lengths.mean()
print(mu)
```

```python
# Calculate the squared differences with respect to the mean
sq_dif = (rand_title_lengths - mu)**2
print(sq_dif)
```

```python
# Calculate the variance (average of the squared differences) and its square root to obtain the standard deviation

variance = sq_dif.mean()

std = variance ** 0.5

print(f"My manual calculation result for Standard Deviation: {std}")
# (/ n - 1) vs (/ n) population, setting ddof=0 for .std()
print(f"Checking this against the .std() function: {rand_title_lengths.std(ddof=0)}")
```
